### There may be other non-numeric values as well, let's see what they are.

### We assume, that any column that have more then 20 unique values, is a continuous variable or column with many names(high cardinality features) ,to show what are the unique values in our features and do not print if there are more then 20 values.

```python
for feature in data.columns:
    
    if data[feature].nunique() <= 20:
        print(feature,data[feature].unique())
```

### From the above we can certainly see some phrases such as  NONE, No info, also there are words that are the same but in upper and lower case, such as 'GLOBAL' and 'GLObaL', etc...

### So we're going to lowercase everything and then try to process any inappropriate phrases,exept of two columns - Company_Name and Dependent-Company Status. First is just an id of a company and the second is our dependant variable.

```python
#We copy our data so as not to change it
dataset = data.copy()
dataset.drop(['Company_Name' ,'Dependent-Company Status'],axis = 1, inplace = True)
```

```python
#Bring every str value to lowercase

for feature in dataset.columns:
    dataset[feature] = dataset[feature].map(lambda x:  str(x).lower().strip())
```

### Now we can easily find out all inapropriate values and replace them to np.nan

### For convenience, we can split our features into categorical(with only srings) and numerical variables.

## If there is at least one number in the column, we add it to the numeric features,because of the phrase no info, the dtype of this column is object, but in fact it is mostly numeric values and several strings.

```python
numerical_features = [] 
for feature in dataset.columns:
    for i in dataset[feature].values:
        if str(i).replace('.',"",1).isdigit() or (isinstance(i,float) and not np.isnan(i)):
            numerical_features.append(feature)
            break
```

```python
dataset[numerical_features].head()
```

```python
#Every feature that not in numerical_features, is a categorical
categorical_features = [feat for feat in dataset.columns if feat not in numerical_features]
dataset[categorical_features].head(3)
```

```python
#Finding all values in numerical features, that should be replaced
to_nan = set()
for col in dataset[numerical_features]:
    for point in dataset[col]:
        if str(point).replace('.',"",1).replace('-','',1).isdigit():
            continue
        else:
            to_nan.add(point)

            
print('Values to be replaced by NaN in numerical variables is',*to_nan)
```

### In categorical columns we have to find that phrases manually

```python
for feature in dataset[categorical_features]:
    
    if dataset[feature].nunique() <= 20:
        
        print(feature,list(dataset[feature].unique()))
```

### And in categorical variables the phrases , that should be replaced by NaN are no info,none.


### Now we can replace all these values

```python
for feature in dataset:
    dataset[feature].replace({'none':np.nan,'no info':np.nan,'unknown amount':np.nan,'nan':np.nan,'not applicable':'no'},inplace = True)
```

```python
# checking for null values in the dataset
Nan_vals = pd.DataFrame(dataset.isnull().sum(),columns=["Nan Values"])
Nan_vals["% of Nan Values"] = (dataset.isna().sum()/len(dataset)*100)
Nan_vals.style.background_gradient(cmap='viridis',low =0.2,high=0.1) 
```

### Now we can drop all columns that have high amount of Nan values. Then we have to bring our numerical variables to its real dtypes(int,float),and also convert categorical features to numerical. 

```python
#Columns with high amount of Nan values
cols_with_many_nans =  Nan_vals.index[np.where(Nan_vals['% of Nan Values'] > 36)]
print('Columns that will be dropped are\n')
for col in cols_with_many_nans:
    print(f'{col} with {round(Nan_vals["% of Nan Values"][col],2)} percent of Nan values')
    dataset.drop(col,axis = 1,inplace =True)
    if col in numerical_features:
        numerical_features.remove(col)
    else:
        categorical_features.remove(col)
```

```python
#bringing all numerical variables to true dtype
for col in dataset:
    try:
        dataset[col] = pd.Series(dataset[col],dtype = 'float64')
    except:
        continue
```

```python
# converting  Est. Founding Date and Last Funding Date from object to datetime
data["Est. Founding Date"] = pd.to_datetime(data["Est. Founding Date"])
data["Last Funding Date"] = pd.to_datetime(data["Last Funding Date"])
```

## Now we want to fill all Nan values, and there are some strategies for doing this.
### 1. Random Sample Imputation
     Random sample imputation consists of taking random observation from the dataset and we use this observation to replace the nan values
    
### 2. End of Distribution imputation
      Filling Nan values with outliers
    
### 3. Arbitrary Value Imputation
      This technique was derived from kaggle competition It consists of replacing NAN by an arbitrary value

### 4. Droping all Nan values
      This will work good, if we have millions of datapoints, and most of thrm is not NaN

### 5. Predicting the missing values with the help of a ML algorithm

### 6. Men,median,mode imputation

## Since we have more than 100 features, we will choose the last method, because it is easy to implement(Robust to outliers),faster way to obtain the complete dataset 

### 1.If the variable is continous then we will replace the missing value by Mean,Median based on the distribution of the variable.
### 2. And if the varibale is categorical then it will be replaced by using Mode of the variable.

## Filing Nan values

```python
#Adding our dependent feature already converted to numeric values
dataset['Success'] = np.array(pd.get_dummies(data['Dependent-Company Status']))[:,1]
```

```python
dataset[categorical_features] = dataset[categorical_features].fillna(dataset[categorical_features].mode().iloc[0])
```

## Now we will divide the features with ordinal relationship and without

```python
ordinal = []
not_ordinal = []
print('Features without ordinal relationships \n')
for i in dataset[categorical_features]:
    if 'yes' not in dataset[i].unique() and 'medium' not in dataset[i].unique() and 'phd' not in dataset[i].unique():
        print(i,dataset[i].unique())
        not_ordinal.append(i)
    else:
        ordinal.append(i)
```

```python
#Replacing the values
for feature in ordinal:
    dataset[feature].replace({'yes':1,'no':0,'low':1, 'medium' :2,'high':3,'small':1,'large':3,'nothing':0, 'bachelors':1,'masters':2,'phd':3},inplace = True)
```

```python
#Dummy features for nonordinal features
dummy = pd.get_dummies(dataset[not_ordinal],drop_first = True)
dataset.drop(not_ordinal,inplace = True, axis = 1)
dataset = pd.concat([dummy,dataset],axis = 1)
```

```python
dataset.head()
```

```python
## Now we will fill nan values of numericl features
dataset[numerical_features] = dataset[numerical_features].fillna(dataset[numerical_features].median().iloc[0])
```

```python
# checking for null values in the dataset
Nan_vals = pd.DataFrame(dataset.isnull().sum(),columns=["Nan Values"])
Nan_vals["% of Nan Values"] = (dataset.isna().sum()/len(dataset)*100)
Nan_vals.style.background_gradient(cmap='viridis',low =0.2,high=0.1) 
```

# Feature Selection 

```python
from sklearn.feature_selection import SelectKBest
from  sklearn.feature_selection import f_classif 

X = dataset.iloc[:,0:-2]  #independent columns
y = dataset.iloc[:,-1]    #target column i.e price range
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=f_classif, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
X = dataset.iloc[:,0:-2]  #independent columns
y = dataset.iloc[:,-1]    #target column i.e price range
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
#print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(12).plot(kind='barh')
plt.show()
```

```python
corr_dict = {}
for feat in dataset:
    if feat != 'Success':
        corr = dataset[feat].corr(dataset['Success'])
        corr_dict[feat] = abs(corr)
    
corr_dict = dict(sorted(corr_dict.items(),key = lambda  x: x[1],reverse=True))
```
